# Task 2

## part 1

plotting the number of trips for each business in each month

```python
import matplotlib.pyplot as plt
import ast

# Read the data from the text file
with open("output/taskTwo_1.txt") as f:
    data_string = f.read()

# Convert the string into a Python list using ast.literal_eval
data = ast.literal_eval(data_string)

plt.figure(figsize=(12,8))
nums = []
names = []
# for accessing the required data for plotting
for item in data:
  nums.append(item[1])
  names.append('{}-{}'.format(item[0][0], item[0][1]))
# plotting
bars = plt.bar(names, nums, width=0.4)
plt.ylabel('trip counts')
plt.xlabel('business-month')
plt.title('number of trips for each business in each month')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x(), yval + .005, '{:.2f}'.format(yval))
```

# 1

plotting average waiting time over time

```python
import matplotlib.pyplot as plt
import ast

# Read the data from the text file
with open("output/taskFive_1.txt") as f:
    data_string = f.read()

# Convert the string into a Python list using ast.literal_eval
data = ast.literal_eval(data_string)
# defining the size of the plot
plt.figure(figsize=(15,8))
days = []
avg_wait = []
# for accessing the required data for plotting
for item in data:
  days.append(item[0])
  avg_wait.append(item[1])
# plotting
bars = plt.bar(days, avg_wait)
plt.xlabel('days')
plt.ylabel("average waiting time")
plt.title("average waiting time over time")
plt.xticks(days)
```
