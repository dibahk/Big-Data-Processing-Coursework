## part 1

plotting the number of trips for each business in each month

```python
import matplotlib.pyplot as plt
import ast

# Read the data from the text file
with open("output/taskTwo_1.txt") as f:
    data_string = f.read()

# Convert the string into a Python list using ast.literal_eval
data = ast.literal_eval(data_string)

plt.figure()
nums = []
names = []
# for accessing the required data for plotting
for item in data:
  nums.append(item[1])
  names.append('{}-{}'.format(item[0][0], item[0][1]))
# plotting
plt.bar(names, nums, width=0.8)
plt.ylabel('trip counts')
plt.xlabel('business-month')
plt.title('number of trips for each business in each month')
```

## Part 3

plotting  driver's earnings (driver_total_pay field) for each business in each month

```python
import matplotlib.pyplot as plt
import ast

# Read the data from the text file
with open("output/taskTwo_3.txt") as f:
    data_string = f.read()

# Convert the string into a Python list using ast.literal_eval
data = ast.literal_eval(data_string)

plt.figure()
nums = []
names = []
# for accessing the required data for plotting
for item in data:
  nums.append(item[1])
  names.append('{}-{}'.format(item[0][0], item[0][1]))
# plotting
bars = plt.bar(names, nums, width=0.8)
plt.ylabel('trip counts')
plt.xlabel("driver's earnings")
plt.title("driver's earnings (driver_total_pay field) for each business in each month")
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x(), yval + .005, '{:.2f}'.format(yval))
```
